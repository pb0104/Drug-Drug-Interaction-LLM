## Load Data ##

```python
# store dictionary with each dataset and the interaction it involves
# we have datasets A, B, D, H, L, P, R, V

dataset_by_interactions = {
    'A': 'Interaction involving alimentary tract and metabolism drugs',
    'B': 'Interaction involving blood and blood forming organs drugs',
    'D': 'Interaction involving dermatologicals drugs',
    'H': 'Interaction involving systemic hormonal preparations, excluding sex hormones and insulins drugs',
    'L': 'Interaction involving antineoplastic and immunomodulating agents drugs',
    'P': 'Interaction involving antiparasitic products, insecticides and repellents drugs',
    'R': 'Interaction involving respiratory system drugs',
    'V': 'Interaction involving various drugs'
}
```

```python


data_folder = os.path.join(project_root, "data")

# dictionary to store dataframes
dataframes = {}

for file_name in os.listdir(data_folder):
    if file_name.endswith(".csv"):
        file_path = os.path.join(data_folder, file_name)
        df_name = os.path.splitext(file_name)[0]
        dataframes[df_name] = pd.read_csv(file_path)
        print(f"Loaded {file_name} into dataframe '{df_name}'")


        df_name = os.path.splitext(file_name)[0]
        df_letter = df_name[-1]

        dataframes[df_name]['Dataset'] = df_letter
        dataframes[df_name]['Category'] = dataset_by_interactions[df_letter]

# make sure all the columns in each dataframe are the same
for df in dataframes.values():
    assert all(df.columns == dataframes[next(iter(dataframes))].columns)
    print(df.columns)

```

```python
# look at first dataset to make sure its loaded correctly
dataframes['ddinter_downloads_code_A'].head()
```

```python
# create a df and combine all of them
df = pd.concat(dataframes.values(), ignore_index=True)
df
```

### Flexibility in Designing Training Question/Answers

```python
# hardcode severity explanations, so it'll include one of these in the answer based on the Interaction Level value
severity_explanations = {
    "Major": "This combination may lead to serious clinical consequences and should generally be avoided or used only with close medical supervision.",
    "Moderate": "This combination may require dose adjustments or additional monitoring to ensure patient safety.",
    "Minor": "This interaction is usually mild but should still be considered in the overall treatment plan.",
    "Unknown": "There is limited or insufficient evidence about the clinical significance of this interaction."
}




def make_qa(row):
    drug_a = row["Drug_A"]
    drug_b = row["Drug_B"]
    level  = row["Level"]
    cat    = row["Category"]

    question = f"How do {drug_a} and {drug_b} interact?"

    # this is where we can change how we format the answer
    explanation = severity_explanations.get(level, "")

    answer = (
        f"{drug_a} and {drug_b} are reported to have a {level.lower()} interaction. "
        f"This interaction is classified under {cat.lower()}. "
        f"{explanation}"
    )

    # answer = (
    #     f"When taken together, {drug_a} and {drug_b} may result in a {level.lower()} interaction. "
    #     f"This type of interaction is related to {cat.lower()}. "
    #     f"{explanation}"
    # )
    return question, answer


data = df.copy()
data[["question", "answer"]] = data.apply(lambda r: pd.Series(make_qa(r)), axis=1)


"""
should we add variation in the question formats for finetuning?
Because you're fine-tuning for general questions like:

"Can I take X and Y together?"

You should add variation in question phrasing.

Otherwise your model will overfit to only:

"How do X and Y interact?"
"""

```
